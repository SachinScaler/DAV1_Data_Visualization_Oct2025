```python
cat_count = df['Genre'].value_counts()
cat_count
```

```python
plt.figure(figsize=(10, 8))
plt.bar(cat_count.index , cat_count.values)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.bar(cat_count.index , cat_count.values)
plt.xticks(rotation=90, fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.bar(cat_count.index , cat_count.values, width=0.2, color = 'orange')
plt.xticks(rotation=90, fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(10, 8))

# manually calculate counts
plt.bar(cat_count.index , cat_count.values, width=0.5, color = 'orange')
plt.title('Games per Genre',fontsize=15)
plt.xlabel('Genre',fontsize=12)
plt.ylabel('Count of Games',fontsize=12)
plt.xticks(rotation=90, fontsize=16)
plt.show()
```

### Pie Chart for Percentage

```python
df.head()
```

```python
sales_data = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
sales_sum = sales_data.sum() # region -> sum of sales
sales_sum # region wise sum of sales
```

```python
plt.figure(figsize=(10, 8))
plt.pie(sales_sum.values,
        labels=sales_sum.index,
        startangle=90,
        explode=(0,0,0.1, 0.2),
        autopct='%.2f%%', # displays percentages
        )
plt.show()
```
